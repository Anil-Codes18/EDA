```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/employee_salary_dataset.csv")
```

```python
sns.barplot(x="Department", y="Monthly_Salary", data=df)
plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Department vs Salary")
plt.show()
```

```python
sns.histplot(df["Monthly_Salary"], bins=5, kde=True)
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Distribution of Monthly Salary")
plt.show()
```

```python
plt.boxplot(df["Monthly_Salary"])
plt.ylabel("Salary")
plt.title("Box Plot of Salary")
plt.show()
```

```python
corr = df[["Experience_Years","Age","Monthly_Salary"]].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix of Employee Data")
plt.show()
```

```python
plt.scatter(df["Experience_Years"], df["Monthly_Salary"])
plt.xlabel("Experience_Years")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
```
